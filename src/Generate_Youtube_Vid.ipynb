```python
import os
from body_matrix import load, export

video_dir = "/Users/troydo42/Desktop/Raw_Vids/"
video_list = os.listdir(video_dir)
video_links = []

for video in video_list:
    file_type = video[-3:]
    if file_type == "mp4":
        print("Adding ", video)
        video_path = video_dir + video
        video_links.append(video_path)
```

```python
video, frames_counts, fps, sample_frame = load.video(
    video_links[0], 
    -90, 
    1
)
```

```python
pil_images = []

for vid in video_links:
    print("Loading video ",vid)
    video, frames_counts, fps, sample_frame = load.video(
        vid, 
        -90, 
        1
    )
    pil_images.append(sample_frame)

print(pil_images)
```
